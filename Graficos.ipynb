```python
import pandas as pd
import matplotlib.pyplot as plt


# Executa uma consulta armazena o resultado em um DataFrame

def gerar_dataframe(connection, consulta):
    query = (consulta)
    df = pd.read_sql(query, connection)
    return df

dataframe1 = pd.read_csv('consulta 1.csv')
dataframe2 = pd.read_csv('consulta 2.csv')
dataframe3 = pd.read_csv('consulta 3.csv')

```

```python
consultor_counts = dataframe1['nome_consultor'].value_counts()
consultor_counts.plot(kind='bar', xlabel='Consultor', ylabel='NÃºmero de Leads convertidos', title='Contagem de Leads por Consultor')
plt.show()
```

```python
pivot_table = dataframe2.pivot_table(index='nome_responsavel', columns='nome_curso', values='total_pagamentos', aggfunc='sum')
pivot_table.plot(kind='bar', stacked=True)
plt.xlabel('ResponsÃ¡vel')
plt.ylabel('Total de Pagamentos')
plt.title('Total de Pagamentos por ResponsÃ¡vel e Curso')
plt.legend(title='Curso', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
```

```python
plt.pie(dataframe3['media_score'], labels=dataframe3['nome_aluno'], autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.axis('equal')  # Garante que o grÃ¡fico de pizza seja desenhado como um cÃ­rculo.
plt.title('MÃ©dia de Score por Aluno e Curso')
```
